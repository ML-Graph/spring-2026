```python
"""
Demonstration: scalar operations with vectors in Python
- Pure Python lists
- NumPy arrays (recommended for real work)
"""

from typing import List

# -----------------------------
# 1) Pure Python (list-based)
# -----------------------------
def scalar_vector_ops_list(s: float, v: List[float]) -> None:
    add = [s + x for x in v]        # scalar + vector (element-wise)
    sub = [x - s for x in v]        # vector - scalar
    mul = [s * x for x in v]        # scalar * vector
    div = [x / s for x in v]        # vector / scalar

    print("== Pure Python lists ==")
    print(f"scalar s = {s}")
    print(f"vector v = {v}")
    print(f"s + v   = {add}")
    print(f"v - s   = {sub}")
    print(f"s * v   = {mul}")
    print(f"v / s   = {div}")
    print()


# -----------------------------
# 2) NumPy (array-based)
# -----------------------------
def scalar_vector_ops_numpy(s: float) -> None:
    import numpy as np

    v = np.array([1.0, -2.0, 3.5, 0.0], dtype=float)

    add = s + v          # broadcasting
    sub = v - s
    mul = s * v
    div = v / s

    # Useful extras
    dot = np.dot(v, v)   # dot product (v · v)
    norm = np.linalg.norm(v)

    print("== NumPy arrays ==")
    print(f"scalar s = {s}")
    print(f"vector v = {v}")
    print(f"s + v   = {add}")
    print(f"v - s   = {sub}")
    print(f"s * v   = {mul}")
    print(f"v / s   = {div}")
    print(f"v · v   = {dot}")
    print(f"||v||   = {norm:.6f}")
    print()


if __name__ == "__main__":
    s = 2.0
    v_list = [1.0, -2.0, 3.5, 0.0]

    scalar_vector_ops_list(s, v_list)
    scalar_vector_ops_numpy(s)
```

```python
"""
Demonstration: scalar operations with a matrix in Python
- Pure Python (list of lists)
- NumPy (recommended)
"""

from typing import List


# --------------------------------
# 1) Pure Python (list of lists)
# --------------------------------
def scalar_matrix_ops_list(s: float, M: List[List[float]]) -> None:
    add = [[s + x for x in row] for row in M]   # scalar + matrix
    sub = [[x - s for x in row] for row in M]   # matrix - scalar
    mul = [[s * x for x in row] for row in M]   # scalar * matrix
    div = [[x / s for x in row] for row in M]   # matrix / scalar

    print("== Pure Python lists ==")
    print(f"scalar s = {s}")
    print("matrix M =")
    for row in M:
        print(row)
    print("s + M =")
    for row in add:
        print(row)
    print("M - s =")
    for row in sub:
        print(row)
    print("s * M =")
    for row in mul:
        print(row)
    print("M / s =")
    for row in div:
        print(row)
    print()


# --------------------------------
# 2) NumPy (array-based)
# --------------------------------
def scalar_matrix_ops_numpy(s: float) -> None:
    import numpy as np

    M = np.array([[1.0, -2.0, 3.0],
                  [4.5,  0.0, -1.5]], dtype=float)

    add = s + M          # broadcasting
    sub = M - s
    mul = s * M
    div = M / s

    # Useful extras
    trace = np.trace(M)
    fro_norm = np.linalg.norm(M, ord="fro")

    print("== NumPy arrays ==")
    print(f"scalar s = {s}")
    print("matrix M =")
    print(M)
    print("s + M =")
    print(add)
    print("M - s =")
    print(sub)
    print("s * M =")
    print(mul)
    print("M / s =")
    print(div)
    print(f"trace(M) = {trace}")
    print(f"Frobenius norm ||M||_F = {fro_norm:.6f}")
    print()


if __name__ == "__main__":
    s = 2.0
    M_list = [
        [1.0, -2.0, 3.0],
        [4.5,  0.0, -1.5]
    ]

    scalar_matrix_ops_list(s, M_list)
    scalar_matrix_ops_numpy(s)
```

## Gradient

```python
import torch

# Define input tensor with gradient tracking enabled
x = torch.tensor(3.0, requires_grad=True)

# Define function y = 2x
y = 2 * x

# Compute derivative dy/dx
y.backward()

# Access gradient
print("x =", x.item())
print("y =", y.item())
print("dy/dx =", x.grad.item())
```

```python
import torch

# x[0] = x1, x[1] = x2
x = torch.tensor([1.5, -0.5], requires_grad=True)

y = 2 * x[0] + 3 * x[1]
y.backward()

print("x1 =", x[0].item(), "x2 =", x[1].item())
print("y  =", y.item())
print("∂y/∂x1 =", x.grad[0].item())
print("∂y/∂x2 =", x.grad[1].item())
```

```python
import torch

def f(x):
    """
    x: tensor of shape (2,) -> [x1, x2]
    returns: tensor of shape (2,) -> [y1, y2]
    """
    y1 = 2 * x[0] + 3 * x[1]
    y2 = 4 * x[0] + 5 * x[1]
    return torch.stack([y1, y2])

# Input
x = torch.tensor([1.0, 2.0], requires_grad=True)

# Compute Jacobian
J = torch.autograd.functional.jacobian(f, x)

print("Jacobian dy/dx:")
print(J)
```

## Paper Classification

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

# Example corpus of paper abstracts (toy dataset)
ml_abstracts = [
    "We use a learning algorithm to model inflation based on training data.",
    "Neural network models predict inflation trends from economic data.",
    "Machine learning improves inflation forecasting with training-based models.",
    "This study applies a predictive learning algorithm to inflation modeling.",
    "Our model learns to predict inflation using neural networks and structured data.",
    "We evaluate algorithmic models that learn inflation patterns from data.",
    "Learning-based inflation models are trained on macroeconomic time series.",
    "The prediction model uses training data and machine learning techniques.",
    "Inflation is modeled using data-driven algorithms with predictive learning.",
    "Our learning system models inflation trends using economic indicators."
]

non_ml_abstracts = [
    "We analyze inflation using an economic theory-driven model and historical data.",
    "Inflation modeling relies on classical economic indicators and structured data.",
    "This study examines inflation dynamics using theoretical models and analysis.",
    "Inflation is analyzed through traditional models and economic frameworks.",
    "Economic data supports a theory-based approach to modeling inflation.",
    "The model explains inflation trends through historical economic data.",
    "A theory-driven framework models inflation based on macroeconomic variables.",
    "We use economic assumptions to analyze inflation using structured models.",
    "Historical data is used to model inflation without predictive algorithms.",
    "Inflation analysis is grounded in economic theory and empirical modeling."
]



texts = ml_abstracts + non_ml_abstracts
labels = [1] * len(ml_abstracts) + [0] * len(non_ml_abstracts)

```

```python
vectorizer = CountVectorizer(stop_words='english', binary=True)
X = vectorizer.fit_transform(texts)
feature_names = vectorizer.get_feature_names_out()
print(f"Number of keyword features: {len(feature_names)}")
print("Sample keywords:", feature_names[:10])
```

```python
import torch
import torch.nn as nn

num_features = X.shape[1]         # vocabulary size (e.g. 80)
num_classes = len(set(labels))    # e.g. 2 for binary classification

model = nn.Linear(num_features, num_classes)
print(model)
# Linear(in_features=80, out_features=2, bias=True)
```

```python
from torch.utils.data import TensorDataset, DataLoader
import torch.optim as optim



# Convert feature matrix and labels to tensors
X_tensor = torch.tensor(X.toarray(), dtype=torch.float32)
y_tensor = torch.tensor(labels, dtype=torch.long)


# Group based on Y and for each group get 5 to get total 10
unique_labels = torch.unique(y_tensor)
train_indices = []
test_indices = []
for lbl in unique_labels:
    lbl_indices = (y_tensor == lbl).nonzero(as_tuple=True)[0]
    train_indices.extend(lbl_indices[:5].tolist())
    test_indices.extend(lbl_indices[5:].tolist())

X_train, X_test = X_tensor[train_indices], X_tensor[test_indices]
y_train, y_test = y_tensor[train_indices], y_tensor[test_indices]




# Define optimizer and loss function
optimizer = optim.SGD(model.parameters(), lr=0.01)
criterion = nn.CrossEntropyLoss()

# Training loop
epochs = 120
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
```

```python
model.eval()
with torch.no_grad():
    logits = model(X_test)
    predictions = torch.argmax(logits, dim=1)
accuracy = (predictions == y_test).float().mean().item()
print(f"Test accuracy: {accuracy:.2f}")
```

```python
# Get the learned weights and bias from the model
weights = model.weight.detach().numpy()  # shape: [num_classes, num_features]
bias = model.bias.detach().numpy()       # shape: [num_classes]

# For binary classification, weights[0] corresponds to class 0 and weights[1] to class 1.
# Let's find the top 5 highest-weight keywords for each class:
vocab = vectorizer.get_feature_names_out()
class0_weights = weights[0]
class1_weights = weights[1]

# Get indices of top positive weights for each class
top5_class0_idx = np.argsort(class0_weights)[-5:]
top5_class1_idx = np.argsort(class1_weights)[-5:]
print("Top keywords for Class 0 (non-ML):", [vocab[i] for i in top5_class0_idx])
print("Top keywords for Class 1 (ML-related):", [vocab[i] for i in top5_class1_idx])
```

## Bayesian Updating

```python
import numpy as np

def posterior_after_n_earthquakes(n: int,
                                 priors=None,
                                 p_survive_given_type=None):
    """
    Compute posterior P(T | survived n earthquakes) for T in {W, G, S}.

    Assumptions:
    - Building type T is fixed.
    - Each earthquake survival event is conditionally independent given T.
    - Observed outcome is 'ND' (not destroyed) for all n earthquakes.

    Returns:
    - dict of posteriors for each type
    """
    if priors is None:
        priors = {"W": 0.3, "G": 0.5, "S": 0.2}
    if p_survive_given_type is None:
        p_survive_given_type = {"W": 0.4, "G": 0.8, "S": 0.95}

    types = ["W", "G", "S"]

    # Unnormalized posterior: P(T) * P(ND|T)^n
    unnorm = {t: priors[t] * (p_survive_given_type[t] ** n) for t in types}

    # Evidence: sum over types
    evidence = sum(unnorm.values())

    # Normalize
    post = {t: unnorm[t] / evidence for t in types}

    return post, evidence, unnorm


if __name__ == "__main__":
    for n in [1, 2, 3, 5, 10]:
        post, evidence, unnorm = posterior_after_n_earthquakes(n)
        print(f"\nAfter n = {n} earthquakes (survived all):")
        print(f"  evidence P(ND^n) = {evidence:.6f}")
        print("  unnormalized scores:")
        for t in ["W", "G", "S"]:
            print(f"    {t}: {unnorm[t]:.6f}")
        print("  posterior P(T | ND^n):")
        for t in ["W", "G", "S"]:
            print(f"    {t}: {post[t]:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

def bayes_update(prior: np.ndarray, likelihood: np.ndarray) -> np.ndarray:
    """posterior ∝ prior * likelihood; normalized to sum to 1"""
    unnorm = prior * likelihood
    return unnorm / unnorm.sum()

# Types and parameters (from your example)
types = ["W", "G", "S"]
prior0 = np.array([0.3, 0.5, 0.2], dtype=float)              # P(T)
p_survive = np.array([0.4, 0.8, 0.95], dtype=float)          # P(ND | T)

# Number of sequential updates (survived each earthquake)
N = 20

# Posterior after each update (k = 0..N)
post = np.zeros((N + 1, len(types)), dtype=float)
post[0] = prior0
for k in range(1, N + 1):
    post[k] = bayes_update(post[k - 1], p_survive)  # observe ND each time

# --------------------------
# Plot 1: Discrete distribution after EACH update (animated bar chart)
# --------------------------
import matplotlib.animation as animation

fig, ax = plt.subplots()
x = np.arange(len(types))
bars = ax.bar(x, post[0])
ax.set_xticks(x)
ax.set_xticklabels(types)
ax.set_ylim(0, 1.0)
ax.set_xlabel("Building Type T")
ax.set_ylabel("Posterior Probability P(T | ND^k)")
title = ax.set_title("Posterior after k = 0 updates (prior)")

def animate(frame):
    probs = post[frame]
    for b, h in zip(bars, probs):
        b.set_height(float(h))
    title.set_text(f"Posterior after k = {frame} updates (survived {frame} earthquakes)")
    return (*bars, title)

ani = animation.FuncAnimation(fig, animate, frames=N + 1, interval=500, blit=False)

# out_path = Path("/posterior_updates.gif")
# ani.save(out_path, writer=animation.PillowWriter(fps=2))

# plt.show()

# --------------------------
# Plot 2: "Visualize the whole" — trajectories of posterior over updates
# --------------------------
plt.figure()
k_vals = np.arange(N + 1)
for i, t in enumerate(types):
    plt.plot(k_vals, post[:, i], label=t)

plt.xlabel("Number of updates k (survived k earthquakes)")
plt.ylabel("Posterior Probability")
plt.title("Posterior trajectories across sequential Bayesian updates")
plt.ylim(0, 1.0)
plt.grid(True)
plt.legend(title="Type")
plt.show()

# --------------------------
# Plot 3: Snapshot distributions at selected iterations (single figure)
# --------------------------
snapshots = [0, 1, 2, 3, 5, 10, 20]
snapshots = [s for s in snapshots if s <= N]

plt.figure()
bar_width = 0.1
for idx, k in enumerate(snapshots):
    plt.bar(x + idx * bar_width, post[k], width=bar_width, label=f"k={k}")

plt.xticks(x + bar_width * (len(snapshots) - 1) / 2, types)
plt.xlabel("Building Type T")
plt.ylabel("Posterior Probability")
plt.title("Posterior distributions at selected updates")
plt.ylim(0, 1.0)
plt.grid(True)
plt.legend(title="Snapshot")
plt.show()

print(f"Saved animation GIF to: {out_path}")
```
